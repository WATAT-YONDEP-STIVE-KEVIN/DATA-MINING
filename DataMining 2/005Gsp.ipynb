## <center>Implementation de l'algorithme GSP</center>

```python
from mlxtend.frequent_patterns import apriori  
from mlxtend.frequent_patterns import association_rules  
import pandas as pd  
import numpy as np
# Define the dataset  
data = {  
    'S.ID': [1, 2, 3, 4, 5],  
    'Sequence': [  
        "{1 5}{2}{3}{4}",  
        "{1}{3}{4}{3 5}",  
        "{1}{2}{3}{4}",  
        "{1}{3}{5}",  
        "{4}{5}"  
    ]  
}  

# Create DataFrame from the dataset  
df = pd.DataFrame(data)  

# Convert sequences into a list of sets (for analysis)  
def preprocess_sequence(sequence):  
    return [set(seq.split()) for seq in sequence.strip("{}").split("}{")]  

df['Processed_Sequence'] = df['Sequence'].apply(preprocess_sequence)  

# Support threshold for GSP  
support_threshold = 0.33 * len(df)  # 33% of total sequences  

# Use the apriori algorithm to find frequent itemsets  
def find_frequent_sequences(df, support_threshold):  
    itemsets = []  
    for index, row in df.iterrows():  
        for seq in row['Processed_Sequence']:  
            itemsets.append(frozenset(seq))  
    itemsets_df = pd.Series(itemsets).value_counts()  
    return itemsets_df[itemsets_df >= support_threshold]  

# Find maximal sequential patterns  
frequent_itemsets = find_frequent_sequences(df, np.round(support_threshold))  
print("Frequent Itemsets with support >= {}: ".format(int(np.round(support_threshold))))  
print(frequent_itemsets)
```
