## <center>Classification hiérachique</center>

```python
import pandas as pd  
import matplotlib.pyplot as plt  
from scipy.cluster.hierarchy import dendrogram, linkage 
```

```python
fromages_data = pd.read_csv("fromages.txt", sep="\t") 
```

```python
data = fromages_data.iloc[:, 1:].values  # Toutes les colonnes sauf la première (Fromages)  
```

```python
# Effectuer la classification hiérarchique  
linked = linkage(data, method='ward') 
```

```python
plt.figure(figsize=(12, 8))  
dendrogram(linked,   
           orientation='top',   
           labels=fromages_data['Fromages'].values,   
           distance_sort='descending',   
           show_leaf_counts=True)  

# Ajouter des titres et des labels  
plt.title('Dendrogramme de Classification Hiérarchique des Fromages')  
plt.xlabel('Fromages')  
plt.ylabel('Distance (Ward)')  
plt.xticks(rotation=90)  # Faire pivoter les étiquettes pour une meilleure lisibilité  
plt.show()
```
