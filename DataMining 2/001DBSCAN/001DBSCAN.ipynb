```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

```python
file_path = "circles.csv"
dataset = pd.read_csv(file_path)
dataset.head(10)
```

```python
print(type(dataset))
```

```python
plt.scatter(dataset.iloc[:,0], dataset.iloc[:,1])
plt.show()
```

```python
plt.hist(dataset.iloc[:,0])
plt.show()
```

```python
plt.hist(dataset.iloc[:,1])
plt.show()
```

```python
dataset.isnull().sum() 
```

```python
ideal_k = []
for i in range(1,21):
    est_kmeans = KMeans(n_clusters=i, random_state=0)
    est_kmeans.fit(dataset)
    ideal_k.append([i, est_kmeans.inertia_])
```

```python
deal_k = np.array(ideal_k)
deal_k
```

```python
ideal_k = np.array(ideal_k)
```

```python
plt.plot(ideal_k[:, 0], ideal_k[:, 1])
plt.show()
```

```python
est_kmeans = KMeans(n_clusters=2, random_state=0)
est_kmeans.fit(dataset)
pred_kmeans = est_kmeans.predict(dataset)
```

```python
plt.scatter(dataset.iloc[:,0], dataset.iloc[:,1], c=pred_kmeans)
plt.show()
```

```python
est_dbscan = DBSCAN(eps=0.1)
pred_dbscan = est_dbscan.fit_predict(dataset)
```

```python
plt.scatter(dataset.iloc[:,0], dataset.iloc[:,1], c=pred_dbscan)
plt.show()
```

```python
est_dbscan = DBSCAN(eps=0.5, min_samples=31) #min_samples = nombre minimum d'chantillons pour former un cluster, eps= distance max entre deux chantillons pour qu'ils soit considres comme etant dans le meme cluster
pred_dbscan = est_dbscan.fit_predict(dataset)
```

```python
plt.scatter(dataset.iloc[:,0], dataset.iloc[:,1], c=pred_dbscan)
plt.show()
```
