## <center>Determination du meilleur k par la methode du courde</center>

```python
import pandas as pd  
import matplotlib.pyplot as plt  
from sklearn.cluster import KMeans  
```

```python
fromages_data = pd.read_csv("fromages.txt", sep="\t")  
```

```python
data = fromages_data.iloc[:, 1:].values  # Toutes les colonnes sauf la première (Fromages)  
```

```python
# Déterminer l'inertie pour différents nombres de clusters (k)  
inertie = []  
k_values = range(1, 11)  # Tester k de 1 à 10  
```

```python
for k in k_values:  
    kmeans = KMeans(n_clusters=k, random_state=42)  # Initialiser KMeans avec k clusters  
    kmeans.fit(data)  # Ajuster le modèle  
    inertie.append(kmeans.inertia_) 
```

```python
# Tracer l'inertie en fonction de k  
plt.figure(figsize=(10, 6))  
plt.plot(k_values, inertie, marker='o')  
plt.title('Méthode du Coude pour Déterminer le Nombre Optimal de Clusters')  
plt.xlabel('Nombre de Clusters (k)')  
plt.ylabel('Inertie')  
plt.xticks(k_values)  
plt.grid()  
plt.axvline(x=3, color='r', linestyle='--')  # Ligne verticale pour choix du k optimal (à ajuster selon l'interprétation)  
plt.show()
```
